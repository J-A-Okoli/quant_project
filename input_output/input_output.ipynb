```python
import pickle 
import numpy as np
import pandas as pd
from random import gauss
import csv
"""
This script demonstrates how to use the pickle module to serialize and deserialize data in Python.
It generates a list of random numbers using the Gaussian distribution,
saves it to a file, and then reads it back into memory.
The pickle module is used for serializing and deserializing Python objects.
Generate a list of random numbers using Gaussian distribution
The gauss feviation 2.
The list comprehension cunction generates a random number from a Gaussian distribution
with mean 1.5 and standard dreates a list of 1 million such random numbers.
"""
a  = [gauss(1.5,2) for i in range(1000000)]

path = "\\Users\\Justi\\Documents"
# opens a file in write-binary mode and writes the data to it 
#data is a list of 1 million random numbers
pkl_file = open(path + 'data.pkl', 'wb')
# using pickle to serialize the data
# and save it to the file
%time pickle.dump(a, pkl_file)
pkl_file.close()
# opens a file in read-binary mode and reads the data from it
pkl_file = open(path + 'data.pkl', 'rb')
# using pickle to deserialize the data 
#deserializing data m,eans loading it back into memory
b = pickle.load(pkl_file)

a[:3], b[:3] # display first 3 elements of both lists

"""Deserialzing and deserializing two objects
The code below demonstrates how to serialize and deserialize two objects using the pickle module."""

pkl_file = open(path + 'data.pkl', 'wb')
#serializing two objects: the list of random numbers and the square of those numbers
#serializing means converting the Python object into a byte stream
#and saving it to a file 
# a byte stream is a sequence of bytes that represents the object
pickle.dump(np.array(a), pkl_file)
pickle.dump(np.array(a) ** 2, pkl_file)
pkl_file.close()

# opens a file in read-binary mode and reads the data from it
pkl_file = open(path + 'data.pkl', 'rb')
# deserializing two objects: the list of random numbers and the square of those numbers
# deserializing means loading the byte stream back into memory
x = pickle.load(pkl_file)   
print(x[:3])
y = pickle.load(pkl_file)
print(y[:3])
# display first 3 elements of both lists
pkl_file.close()

# pickle comtains no metainformation about the objects it serializes
pkl_file = open(path + 'data.pkl', 'wb')
# a workaround is to use dict objects containing all the other objects
pickle.dump({'x': x, 'y': y}, pkl_file)
# dict objects are used to store key-value pairs
pkl_file.close()
# opens a file in read-binary mode and reads the data from it
pkl_file = open(path + 'data.pkl', 'rb')
# deserializing the dict object
data = pickle.load(pkl_file)
#closing the file after reading
pkl_file.close()
# display the keys and first 4 elements of each key's value
for key in data.keys():
    print(key, data[key][:4])  # display first 4 elements of each key's value
"""The following code creates a dummy data set as an ndarray object, creates a Dateti
meIndex object, combines the two, and stores the data as a CSV text file:"""
rows = 5000
# create a dummy data set as an ndarray object
a = np.random.standard_normal((rows, 5)).round(2)
# create a DateTimeIndex object
t = pd.date_range('2020-01-01', periods=rows, freq='h')
print(len(t))

csv_file = open(path + 'data.csv', 'w')
# create a DataFrame object from the dummy data set and DateTimeIndex object
header  = 'date,no1,no2,no3,no4,no5\n'
csv_file.write(header)
# write the data to the CSV file
# the zip function is used to combine the two lists into a list of tuples
# each tuple contains a date and the corresponding values from the dummy data set
for t_, (no1, no2, no3, no4, no5) in zip(t, a):
    s = f"{t_}, {no1},{no2},{no3},{no4},{no5}\n"
    csv_file.write(s)

csv_file.close()
# read the CSV file and read its content
csv_file = open(path + 'data.csv', 'r')
# display the first 5 lines of the CSV file
# using readline to read the first 5 lines
for i in range(5):
    print(csv_file.readline(), end="")  # display first 5 lines of the CSV file

csv_file.close()

csv_file = open(path + 'data.csv', 'r')

content = csv_file.readlines()
# display the first 5 lines of the CSV file
print(content[:5])  # display first 5 lines of the CSV file
csv_file.close()
# read the CSV file using the csv module
# with command is used to open the file and automatically close it after reading
with open(path + 'data.csv', 'r') as csv_file:
    # read the CSV file and read its content
    cvs_reader = csv.reader(csv_file)
    # list comprehension is used to read all lines from the CSV file
    # cvs_reader is an iterator that reads the CSV file line by line
    lines = [line for line in cvs_reader]
print(lines[:5])  # display first 5 lines of the CSV file


with open(path + 'data.csv', 'r') as f:
    # read the CSV file and read its content
    cvs_reader = csv.DictReader(f)
    # list comprehension is used to read all lines from the CSV file
    # cvs_reader is an iterator that reads the CSV file line by line
    lines = [line for line in cvs_reader]
print(lines[:5])  # display first 5 lines of the CSV file

```
